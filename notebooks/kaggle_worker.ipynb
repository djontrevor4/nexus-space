```python
!pip install requests -q
import requests, json, socket, subprocess

VPS = 'http://176.123.169.38:5000'

def get_task():
    try:
        r = requests.get(f'{VPS}/catch/kaggle-task', timeout=10)
        return r.json()
    except: return None

def send_result(data):
    requests.post(f'{VPS}/catch/kaggle-result', json=data, timeout=10)

# Nuclei scanner
!wget -q https://github.com/projectdiscovery/nuclei/releases/download/v3.1.0/nuclei_3.1.0_linux_amd64.zip
!unzip -q -o nuclei_3.1.0_linux_amd64.zip
!chmod +x nuclei

def nuclei_scan(target):
    result = subprocess.run(['./nuclei', '-u', target, '-silent', '-json'],
                          capture_output=True, text=True, timeout=300)
    findings = []
    for line in result.stdout.split('\n'):
        if line.strip():
            try: findings.append(json.loads(line))
            except: pass
    return findings

# Main loop
task = get_task()
if task and task.get('target'):
    print(f'Scanning: {task["target"]}')
    results = nuclei_scan(task['target'])
    send_result({'target': task['target'], 'nuclei': results})
    print(f'Found {len(results)} vulnerabilities')
else:
    print('No tasks. Run manually:')
    print('nuclei_scan("target.com")')
```
